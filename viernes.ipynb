```python
#Medidas de tendencia central
print(f"Promedio: {df['budget_x'].mean()}")
print(f"Mediana: {df['budget_x'].median()}")
print(f"Moda: {df['budget_x'].mode()}")
#Medidad de dispersión
print(f"Desviación estándar: {df['budget_x'].std()}")
print(f"Varianza: {df['budget_x'].var()}")
#Medidas de posición
print(f"Minimo: {df['budget_x'].min()}")
print(f"Q1: {df['budget_x'].quantile(0.25)}")
print(f"Q2: {df['budget_x'].quantile(0.50)}")
print(f"Q3: {df['budget_x'].quantile(0.75)}")
print(f"Maximo: {df['budget_x'].max()}")
```

```python
#Medidas de tendencia central
print(f"Promedio: {df['revenue'].mean()}")
print(f"Mediana: {df['revenue'].median()}")
print(f"Moda: {df['revenue'].mode()}")
#Medidad de dispersión
print(f"Desviación estándar: {df['revenue'].std()}")
print(f"Varianza: {df['revenue'].var()}")
#Medidas de posición
print(f"Minimo: {df['revenue'].min()}")
print(f"Q1: {df['revenue'].quantile(0.25)}")
print(f"Q2: {df['revenue'].quantile(0.50)}")
print(f"Q3: {df['revenue'].quantile(0.75)}")
print(f"Maximo: {df['revenue'].max()}")
```

```python
#SI ENCUENTRAS UNA COLUMNA QUE SE DETECTO COMO TEXTO (OBJECT) Y DEBERIA SER NUMERICA
df["date_x"] = pd.to_numeric(df["date_x"].str.replace("/",""))
```

```python
columnas_numericas = []
columnas_no_numericas = []
for col in df.columns:
    if df[col].dtype == "object":
        columnas_no_numericas.append(col)
    else:
        columnas_numericas.append(col)   
print(f"Columnas numericas: {columnas_numericas}")
print(f"Columnas no numericas: {columnas_no_numericas}")
```

```python
df.groupby("names")["revenue"].mean().reset_index().nlargest(10, "revenue").plot.bar(x="names", y="revenue")
```

```python
df.groupby("genre")["revenue"].mean().reset_index().nlargest(10, "revenue").plot.bar(x="genre", y="revenue")
```

```python
genres = df["genre"].str.replace(" ", "").str.lower().str.get_dummies(sep=",")
df[genres.columns] = genres
df.head()
```

```python
df.melt(id_vars=["names", "revenue"], value_vars=genres.columns, var_name="genre").groupby("genre")["revenue"].mean()
```

```python
df.melt(id_vars=["names", "revenue"], value_vars=genres.columns, var_name="genre").groupby("genre")["revenue"].mean().reset_index().nlargest(10, "revenue").plot.bar(x="genre", y="revenue")
```

```python
df.groupby("country")["revenue"].mean().reset_index().nlargest(20, "revenue").plot.bar(x="country", y="revenue")
```
